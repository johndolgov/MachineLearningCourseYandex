```python
def name_scratcher(data):
    result = []
    for lst in data:
        if('Miss.' in lst):
            index = lst.index('Miss.') + 1
            lst[index]=lst[index].strip('(')
            lst[index]=lst[index].strip(')')
            result.append(lst[index])
        elif ('Mrs.' in lst):
            index = lst.index('Mrs.') + 1
            lst[index]=lst[index].strip('(')
            lst[index]=lst[index].strip(')')
            result.append(lst[index])
        else:
            index = 1;
            lst[index]=lst[index].strip('(')
            lst[index]=lst[index].strip(')')
            result.append(lst[index])
    return result
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import scale
```

```python
wines = pd.read_csv('wine.csv')
wines.head(10)
```

```python
kf = KFold(n_splits=5, random_state=42 ,shuffle=True)
```

```python
for train_index, test_index in kf.split(wines):
    wines_train = wines.iloc[train_index, :]
    wines_test = wines.iloc[test_index, :]
    classes_train = list(wines_train['Class'])
    params_train = wines_train.loc[:,wines_train.columns != 'Class'].values.tolist()
    classes_test = list(wines_test['Class'])
    params_test = wines_test.loc[:,wines_test.columns != 'Class'].values.tolist()
    knn = KNeighborsClassifier(n_neighbors=3)
    knn.fit(params_train, classes_train)
```

```python
classes = list(wines['Class'])
params = wines.loc[:, wines.columns != 'Class'].values.tolist()
params = scale(params)
quality = []
for n in range(1,51):
    knn = KNeighborsClassifier(n_neighbors=n)
    value_list=cross_val_score(knn, params, classes, cv=kf)
    quality.append(value_list.mean())
print(f'optim_k = {quality.index(max(quality)) + 1}, optim_value = {max(quality)}')
```
